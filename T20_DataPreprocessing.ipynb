```python
#import necessary libraries

import pandas as pd

import json
```

## <font color=red>1. Process Match Results</font>

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    
    data = json.load(f)

data
```

```python
data[0]['matchSummary']
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])

df_match.head()
```

```python
df_match.shape
```

**Use scorecard as a match id to link with other tables**

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)

df_match.head()
```

**Create a match ids dictionary that maps team names to a unique match id. This will be useful later on to link with other tables**

```python
match_ids_dict = {}

# This line starts a loop that iterates over each row in the df_match DataFrame. 
# The iterrows() function returns an iterator that yields index and row data.

for index, row in df_match.iterrows():
    
    key1 = row['team1'] + ' Vs ' + row['team2'] # This line creates the first key for the dictionary by concatenating the 
                                                # values of the 'team1' and 'team2' columns from the current row. The key 
                                                # is in the format "team1 Vs team2".
    
    key2 = row['team2'] + ' Vs ' + row['team1'] # This line creates the second key for the dictionary by concatenating the 
                                                # values of the 'team2' and 'team1' columns from the current row. The key 
                                                # is in the format "team2 Vs team1".
    
    match_ids_dict[key1] = row['match_id']      # This line assigns the match ID value from the current row to the first key 
                                                # in the dictionary.
    
    match_ids_dict[key2] = row['match_id']      # This line assigns the match ID value from the current row to the second key 
                                                # in the dictionary.
```

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv', index = False)
```

## <font color=red>2. Process Batting Summary</font>

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    
    data = json.load(f)
    
    all_records = [] # Here, an empty list all_records is created to store the batting summary data.
    
    for rec in data: # The code then iterates over each record in the data variable, which represents the JSON data.
        
        all_records.extend(rec['battingSummary']) # For each record, the batting summary data is extracted using the 
                                                  # key 'battingSummary' and appended to the all_records list using 
                                                  # the extend() function.

print(df_batting.shape)       

df_batting = pd.DataFrame(all_records)

df_batting.head(10)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)

df_batting.head()
```

## <font color=red>3. Process Bowling Summary</font>

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    
    data = json.load(f)
    
    all_records = []
    
    for rec in data:
        
        all_records.extend(rec['bowlingSummary'])

all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)

print(df_bowling.shape)

df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)

df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```
